#**Imports**

```python
import networkx as nx
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import time
```

#**Criacao do Grafo**

```python
def random_edge(graph):
    edges = list(graph.edges)
    nonedges = list(nx.non_edges(graph))

    # random edge choice
    chosen_edge = rd.choice(edges)
    chosen_nonedge = rd.choice([x for x in nonedges if chosen_edge[0] == x[0]])

    # add new edge
    graph.add_edge(chosen_nonedge[0], chosen_nonedge[1])

    return graph
```

```python
num_nodes = 10

graph_10 = nx.Graph()
h = nx.path_graph(num_nodes)
graph_10.add_nodes_from(h)
graph_10.add_edges_from(h.edges())
for i in range(3):
  graph_10 = random_edge(graph_10)

num_nodes = 100
graph_100 = nx.Graph()
h = nx.path_graph(num_nodes)
graph_100.add_nodes_from(h)
graph_100.add_edges_from(h.edges())
for i in range(21):
  graph_100 = random_edge(graph_100)


num_nodes = 1000
graph_1000 = nx.Graph()
h = nx.path_graph(num_nodes)
graph_1000.add_nodes_from(h)
graph_1000.add_edges_from(h.edges())
for i in range(151):
  graph_1000 = random_edge(graph_1000)
```

#**Colocarao do Grafo**

##Busca em Largura

```python
def bfs(graph):
  d_bfs = nx.coloring.greedy_color(graph, strategy="connected_sequential_bfs")
  return d_bfs
```

##Busca em Profundidade

```python
def dfs(graph):
  d_dfs = nx.coloring.greedy_color(graph, strategy="connected_sequential_dfs")
  return d_dfs
```

##DSatur

```python
def dsatur(graph):
  d_ds = nx.coloring.greedy_color(graph, strategy="DSATUR")
  return d_ds
```

##Welsh Powell

```python
def wp(graph):
  d_wp = nx.coloring.greedy_color(graph, strategy="largest_first")
  return d_wp
```

##Timer

```python
storage = []
```

###Grafo 10

```python
tempos_10 = []
temp_storage = []
color_storage = []

start = time.monotonic()
d_bfs_10 = bfs(graph_10).values()
end = time.monotonic()
tempos_10.append(end - start)
d_bfs_10 = np.array(list(d_bfs_10))
temp_storage.append(round(end - start, 4))
color_storage.append(np.unique(d_bfs_10))


start = time.monotonic()
d_dfs_10 = dfs(graph_10).values()
end = time.monotonic()
tempos_10.append(end - start)
d_dfs_10 = np.array(list(d_dfs_10))
temp_storage.append(round(end - start, 4))
color_storage.append(np.unique(d_dfs_10))


start = time.monotonic()
d_ds_10 = dsatur(graph_10).values()
end = time.monotonic()
tempos_10.append(end - start)
d_ds_10 = np.array(list(d_ds_10))
temp_storage.append(round(end - start, 4))
color_storage.append(np.unique(d_ds_10))


start = time.monotonic()
d_wp_10 = wp(graph_10).values()
end = time.monotonic()
tempos_10.append(end - start)
d_wp_10 = np.array(list(d_wp_10))
temp_storage.append(round(end - start, 4))
color_storage.append(np.unique(d_wp_10))

storage.append([temp_storage, color_storage])
```

#Resultados

```python
iterate = 0
for grafo in storage:
  print("= = = = = = = = = = = = = = = = = = = =")
  for i in range(len(grafo[0])):
    if (iterate == 0):
      print("Grafo 10")
    elif (iterate == 1):
      print("Grafo 100")
    elif (iterate == 2):
      print("Grafo 1000")

    if (i == 0):
      print("Duracao BFS:", grafo[0][i])
      print("Cores utilizadas: ", grafo[1][i],  "\n")
    elif (i == 1):
      print("Duracao DFS:", grafo[0][i])
      print("Cores utilizadas: ", grafo[1][i],  "\n")
    elif (i == 2):
      print("Duracao DSatur:", grafo[0][i])
      print("Cores utilizadas: ", grafo[1][i],  "\n")
    elif (i == 3):
      print("Duracao Welsh Powell:", grafo[0][i])
      print("Cores utilizadas: ", grafo[1][i],  "\n\n")
  iterate += 1

tamanhos = [10, 100, 1000]

algoritmo1 = [tempos_10[0], tempos_100[0], tempos_1000[0]]
algoritmo2 = [tempos_10[1], tempos_100[1], tempos_1000[1]]
algoritmo3 = [tempos_10[2], tempos_100[2], tempos_1000[2]]
algoritmo4 = [tempos_10[3], tempos_100[3], tempos_1000[3]]


plt.plot(algoritmo1, tamanhos, marker='o', label='BFS')
plt.plot(algoritmo2, tamanhos, marker='o', label='DFS')
plt.plot(algoritmo3, tamanhos, marker='o', label='DSatur')
plt.plot(algoritmo4, tamanhos, marker='o', label='Welsh Powell')

plt.ylabel('Tamanho do Grafo')
plt.xlabel('Tempo (segundos)')
plt.title('Comparação de Tempo por Tamanho do Grafo')
plt.xlim(0, 0.05)
plt.legend()
plt.show()
```

```python
dict_bfs = nx.coloring.greedy_color(graph_10, strategy="connected_sequential_bfs")
dict_dfs = nx.coloring.greedy_color(graph_10, strategy="connected_sequential_dfs")
dict_ds = nx.coloring.greedy_color(graph_10, strategy="DSATUR")
dict_wp = nx.coloring.greedy_color(graph_10, strategy="largest_first")

dicts = [dict_bfs, dict_dfs, dict_ds, dict_wp]

for d in dicts:
  chaves = list(d.keys())
  chaves.sort()
  sorted_dict = {i: d[i] for i in chaves}
  nx.draw(graph_10, node_color=list(sorted_dict.values()), with_labels=True)
  plt.show()
```
